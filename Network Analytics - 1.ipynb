1. Use NumPy to simulate two, 1-mode, unweighted, and undirected networks with 100 nodes. A first network G_ld should have average degree = 3, while a second network G_hd should have average degree = 8.

```python

# import the library
import matplotlib.pyplot as plt
import networkx as nx

plt.figure(figsize=(12,8))
# initialize a new graph object
G = nx.Graph()

# define the set of nodes
node_list = list(range(0,100))

# populate the graph with nodes
G.add_nodes_from(node_list)

# define the set of edges
edge_list = []
for i in range(50):
    np.random.seed(i)
    a = i
    b = np.random.randint(i+1,96)
    c = np.random.randint(b+1,97)
    d = np.random.randint(c+1,98)
    e = np.random.randint(d+1,99)
    f = np.random.randint(e+1,100)
    g = np.random.randint(f+1,101)
    for letter in [b,c,d,e,f,g]:
        edge = (a,letter)
        edge_list.append(edge) 

# populate the graph with edges
G.add_edges_from(edge_list)

# fix the drawing options
options = {'node_color': 'purple',
           'alpha': 1,
           'node_size': 800,
           'width': 1,
           'with_labels': True,
           'font_size': 24,
           'font_color': 'white'}

# draw the network
nx.draw(G, **options)

# add  title to the network
plt.title('Visual representation\n', fontsize=32)

# show the plot
plt.show()
```
